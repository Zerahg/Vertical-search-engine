```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
import numpy as np
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import wordnet
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer 
import string
```

```python
f = csv.writer(open('saveFile.csv', 'w', encoding='utf-8'))
f.writerow(['authors', 'publication year', 'title', 'links', 'pureportal'])

def extract(page):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.107 Safari/537.36'}
    url = f'https://pureportal.coventry.ac.uk/en/organisations/faculty-of-engineering-environment-computing/publications/?page={page}'
    r = requests.get(url, headers)
    soup = BeautifulSoup(r.content, 'html.parser')
    return soup

def transform(soup):
    lis = soup.find_all('li', class_ = 'list-result-item')
    for item in lis:
        title = item.find('h3').text.strip()

        try:
            pureportal = item.find('a', class_ = 'link person').get('href')
        except:
            pureportal = None
        
        
        year = item.find('span', class_='date').text.strip()
        links = item.find('a', class_='link').get('href', None)
        try:
            authors = item.find('a', class_='link person').text.strip()
        except:
            authors = ''

        f.writerow([authors, year, title, links, pureportal]) # saving the files 

for i in range(0, 96, 1):
    print(f'Getting page, {i}')
    c= extract(0)
    transform(c)
```

```python
# this function is created to convert all the text to lower text and to remove punctuation to avoid it to not be case sensitve when searching
def process_string(text):
  text = text.lower() #to lowercase
  text = text.translate(str.maketrans('','', string.punctuation)) #strip punctuation
  return text
```

```python
# function to tag first charater lemmatize accepts
def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN)
```

```python
stop = stopwords.words('english')
```

```python
# Lemmatization is the process of grouping together the different inflected forms of a word so they can be analysed as a single item
lem = WordNetLemmatizer()

# this function loops through the words by property with the use of a vocabulary and morphological analysis of words, normally aiming to remove inflectional endings only and to return the base or dictionary form of a word
def stop_lemmatize(doc):
    tokens = nltk.word_tokenize(doc)
    tmp = ""
    for w in tokens:
        if w not in stop:
            tmp += lem.lemmatize(w, get_wordnet_pos(w)) + " "
    return tmp

stop_lemmatize(doc = entry.title)
```

```python
# a function to transform a columns in the dataframe and return the end resut which has been processed
def transform_df(df):
    df = df
    df['authors'] = df['authors'].apply(process_string)
    df['title'] = df['title'].apply(process_string)
    df['text'] = df['authors'] + " " + df['title']
    drop_cols = ['authors',  'title', 'publication year']
    df = df.drop(drop_cols, axis=1)
    return df
```

```python
def index_it(entry, index):
  words = entry.text.split()
  ID = entry.ID
  for word in words:
    if word in index.keys():
      if ID not in index[word]:
        index[word].append(ID)
    else:
      index[word] = [ID]
  return index
```

```python
def index_all(df, index):
  for i in range(len(df)):
    entry = df.loc[i, :]
    index = index_it(entry, index)
  return index
```

```python
def build_index(df, index):
    to_add = transform_df(df)
    index = index_all(df = to_add, index = index)
    return index
```
